```python

import numpy as np

# Input data (AND gate)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Output data (AND gate result)
y_and = np.array([[0], [0], [0], [1]])
y_or = np.array([[0], [1], [1], [1]])

# Initialize weights and biases randomly
np.random.seed(1)
w1 = np.random.rand(2, 1)  # weights for the two inputs
b1 = np.random.rand(1)  # bias term
w2 = np.random.rand(2, 1)  # weights for the two inputs
b2 = np.random.rand(1)  # bias term

# Learning rate
learning_rate = 0.1 

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Training the neural network for AND gate
# used w1, b1   
for epoch in range(10000):  
    # Forward pass
    input_layer = X
    and_outputs = sigmoid(np.dot(input_layer, w1) + b1)

    # Calculate the error
    err1 = y_and - and_outputs

    # Backpropagation
    adjustments = err1 * sigmoid_derivative(and_outputs)
    w1 += np.dot(input_layer.T, adjustments) * learning_rate
    b1 += np.sum(adjustments) * learning_rate 

    # Print the error every 1000 epochs
    if epoch % 1000 == 0:
        print(f'Epoch {epoch}, Error: {np.mean(np.abs(err1))}')
        
# Testing the neural network for AND gate
print("Final And Gate outputs after training:")
print("\nFinal weights and bias:")
print("Weights:", w1)
print("Bias:", b1)

print("\nPredictions after training:")

predictions = sigmoid(np.dot(X, w1) + b1)
print(predictions.round())

# Training the neural network for OR gate
#used w2, b2
for epoch in range(10000):  
    # Forward pass
    input_layer = X
    or_outputs = sigmoid(np.dot(input_layer, w2) + b2)

    # Calculate the error
    err2 = y_or - or_outputs

    # Backpropagation
    adjustments = err2 * sigmoid_derivative(or_outputs)
    w2 += np.dot(input_layer.T, adjustments) * learning_rate
    b2 += np.sum(adjustments) * learning_rate 

    # Print the error every 1000 epochs
    if epoch % 1000 == 0:
        print(f'Epoch {epoch}, Error: {np.mean(np.abs(err2))}')  

# Testing the neural network for OR gate
print("Final Or Gate outputs after training:")  
print("\nFinal weights and bias:")
print("Weights:", w2)   
print("Bias:", b2)
print("\nPredictions after training:")
predictions = sigmoid(np.dot(X, w2) + b2)
print(predictions.round())

```
